```python
def isValidRow(row):
    cnt=0
    for cell in row:
        if(cell['width'] < 1.00):
            cnt+=1
    print("avg=",avg,"cnt=", cnt)
    if(cnt<=avg):
        return True
    else:
        return False
```

```python
def nextRowData(nextRow, top, height, left, width, varWidth):
    #print("Cell data= ",top,height, left, width)
    bottom = top+height
    bottom = bottom - (bottom*0.05)
    
    End = left+width
    End = End+(End*0.05)

    nextTop = nextRow[0]
    nextTop = nextTop['top']
    #print(nextTop, bottom)

    if(nextTop < bottom):
        #print("Next row= ", nextRow)
        #print(left)
        leftR = left-(left*0.05)
        for cell in nextRow:
            cellEnd = cell['left']+cell['width']
            #print(cell['left'], leftR,"  ",End, cellEnd,"  ",cell['width'], varWidth)
            if(cell['left']>=leftR and cellEnd <= End and cell['width']>=varWidth and cell['text']!=''):
                #print("Found")
                #print(cell['text'])
                return cell['text']
        return -1
    else:
        return -1
```
