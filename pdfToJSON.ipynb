```python
import pandas as pd
import tabula
import json
```

```python
# Read pdf into DataFrame
df = tabula.read_pdf("Bates.pdf", pages='all',lattice=True, output_format='json')
```

```python
# Convert list to JSON
json_data = json.dumps(df)
json_data
```

```python
# Convert JSON string to dictionary
dict_data = json.loads(json_data)
```

```python
# Load the data from the JSON
data = []
for page in dict_data:
    for row in page['data']:
        for cell in row:
            data.append({'text': cell['text'], 'top': cell['top'], 'left': cell['left'], 'width': cell['width'], 'height': cell['height']})
        #data.append({'text': row['text'], 'top': row['top'], 'left': row['left'], 'width': row['width'], 'height': row['height']})
```
